```python
import numpy as np
import pandas as pd

#Our datas does not have column names. So we add it.
colnames = ['Pregnancy',
    'Glukosis Concentration',
    'Blood Pressure',
    'Triceps Fold Thickness',
    'Serum Insulin',
    'Body Mass Index',
    'Diabetes Family Tree Function',
    'Age',
    'Class Variable']
data = pd.read_csv("veriseti6.csv", names=colnames)
data
"""
#Check the total data number and null values.
data.size
data.isnull().sum()
#We don't have any null values and have 6912 total values."""
```

```python
#The Class Variable indicates us that the Patient has Diabetes or not. So we declare all column datas as X, and declare Class Variable for Y.
X = data.iloc[:,0:-1].values
Y = data.iloc[:,-1].values
```

```python
#We don't have any string value so we don't need any encoder like LabelEncoder or OneHotEncoder. So we continue.
from sklearn.model_selection import train_test_split
Xtr, Xte, Ytr, Yte = train_test_split(X,Y,test_size=0.25, random_state=13)
```

```python
#There are gaps in value between the values of our data. So we fit the data with StandardScaler
from sklearn.preprocessing import StandardScaler

SC = StandardScaler()
Xtr = SC.fit_transform(Xtr)
Xte = SC.fit_transform(Xte)
```

```python
#Build the Neural Network
from keras.models import Sequential
from keras.layers import Dense

#Call Network Builder
Network = Sequential()

#Input Layer
Network.add(Dense(units=128, kernel_initializer='uniform', activation='relu', input_dim=8))

#Hidden Layers
Network.add(Dense(units=128, kernel_initializer='uniform', activation='relu'))
Network.add(Dense(units=128, kernel_initializer='uniform', activation='relu'))

#Our output will be 0 or 1. So we use 1 neuron, and our activation function will be 'sigmoid' for output layer
Network.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

#Compile the Network
Network.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
Network.fit(Xtr, Ytr, epochs=200, validation_data=(Xtr, Ytr))
```

```python
#Predict the data with confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
Pred = Network.predict(Xte)
y_pred = np.where(Pred > 0.5, 1, 0)
CM = confusion_matrix(Yte, y_pred)
print(classification_report(Yte, y_pred))
#We have %70 of accuracy of f1-score, which is performance of binary classification.
```
